Scrape wikipedia

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df=pd.read_html(url,skiprows=0)[0]
df
```

Dropping the rows where Borough is not assigned

```python
df=df[df.Borough !='Not assigned']
df
```

```python
df1=df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
df1
```

```python
df1['Neighbourhood'].replace('Not asssigned','QueensPark',inplace=True)
df1
```

```python
df1.shape
```

```python
data='https://cocl.us/Geospatial_data'
df_geo=pd.read_csv(data)
df_geo
```

```python
df_geo.rename(columns={'Postal Code':'Postcode'},inplace=True)
df_geo.head()
```

```python
df3=pd.merge(df1,df_geo, on=['Postcode'],how='inner')
df3
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Tornoto City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighbourhood in zip(df3['Latitude'], df3['Longitude'], df3['Borough'], df3['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='black',
        fill=True,
        fill_color='#3100cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
